```python
import pandas_datareader.data as web
import pandas as pd
import datetime as dt

codes = ['005930']
df = {}

for code in codes:
    df['Close'] = web.DataReader(code, 'naver', start='2016-12-01', end='2023-07-09')['Close']

df = web.DataReader(code, 'naver', start='2023-01-01', end='2023-07-09')
df = pd.DataFrame(df)

for column in df.columns:
    df[column] = df[column].astype(float)

# df['Open'] = df['High'] = df['Low'] = df['Close'] #종가데이터만 있을경우 실행

df.index.name = None
SAMSUNG = df
SAMSUNG
```

```python
from backtesting import Backtest, Strategy
from backtesting.lib import crossover

from backtesting.test import SMA


class MonthlyBuyStrategy(Strategy):
    def init(self):
        # 월별 초기화 변수
        self.month = None

    def next(self):
        # 현재 날짜의 월을 확인
        current_month = self.data.index[-1].month

        # 월이 변경되었을 때 매수 신호 생성
        if self.month is None or current_month != self.month:
            self.month = current_month
            self.buy()


bt = Backtest(SAMSUNG, MonthlyBuyStrategy,
              cash=1000000,
              trade_on_close = True,
              exclusive_orders=True)

output = bt.run()
bt.plot()
```

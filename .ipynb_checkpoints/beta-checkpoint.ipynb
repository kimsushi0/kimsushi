```python
import pandas_datareader.data as web
import pandas as pd
import datetime as dt

codes = ['005930', 'KOSPI','KOSDAQ']
df = {}

for code in codes:
    df[code] = web.DataReader(code, 'naver', start='2019-09-10', end='2019-10-09')['Close']
df = pd.DataFrame(df)
for column in df.columns:
    df[column] = df[column].astype(float)
df = df.pct_change().dropna()
df
```
